## Roots of an equation

Finding a root of an equation means finding the value where equation is zero.

To do this numerically, we can use 3 Methods

- 1. [Bisection Method](#bisection-method)
- 2. [False-Position Method](#false)
- 3. [Newton-Raphson Method](#nrm)

### Bisection Method

```python
#function
def f(x):
    return x**2-4
```

```python
#Bisection
def bisection(iterations=100, acc=0.0):
    x1=float(input("\nEnter the First Guess: "))
    x2=float(input("Enter the Second Guess: "))
    
    if (f(x1)*f(x2)<0):
        for i in range(iterations):
            x3 = (x1+x2)/2
            if f(x3)*f(x1)<0:
                x2=x3
            else:
                x1=x3
            print(f"{i}",f": Root is {x3} and f(x3) is {f(x3)}")

            if(i<=iterations and abs(f(x3))<acc):
                break
        print(f"\nRoot is {x3} and f(x3) is {f(x3)}")
    else:
        print(f"\nRoot does not lie in given guess \nX1: {x1} & X2: {x2})")
```

```python
def falsi(iterations=100, acc=0.0):
    x1=float(input("\nEnter the First Guess: "))
    x2=float(input("Enter the Second Guess: "))
    
    if (f(x1)*f(x2)<0):
        for i in range(iterations):
            x3 = x1-f(x1)*((x1-x2)/(f(x1)-f(x2)))
        if f(x3)*f(x1)<0:
            x2=x3
        else:
            x1=x3
            print(f"{i}",f": Root is {x3} and f(x3) is {f(x3)}")

            if(i<=iterations and abs(f(x3))<acc):
                break
        print(f"\nRoot is {x3} and f(x3) is {f(x3)}")
    else:
        print(f"\nRoot does not lie in given guess \nX1: {x1} & X2: {x2})")
```
